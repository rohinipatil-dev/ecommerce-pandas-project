```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load cleaned data
df = pd.read_csv("../outputs/cleaned_data.csv")
df["InvoicedDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# Preview
df.head()
```

```python
# Top-Selling Products by Revenue

top_products = df.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar", color="skyblue")
plt.title("Top 10 Products by Revenue")
plt.ylabel("Total Revenue")
plt.xlabel("Products")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("../outputs/visuals/top_products.png")
plt.show()
```

```python
# Monthly Revenue Trend

# Convert InvoiceDate to datetime (if not already done)
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
# Extract Month period
df["InvoiceMonth"] = df["InvoiceDate"].dt.to_period("M")

# Preview
df["InvoiceMonth"].head()

monthly_revenue = df.groupby("InvoiceMonth")["TotalPrice"].sum()

# Preview
monthly_revenue.head()

monthly_revenue.plot(marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("../outputs/visuals/monthly_revenue_trend.png")
plt.show()
```
